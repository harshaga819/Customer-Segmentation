##### 6.1.1 Applying Z-Score

On Profit column Z-Score is applied because it is normally distributed

```python
df['Z-Score']= (df['Profit']-df['Profit'].mean())/(df['Profit'].std()) 
```

```python
Higher_limit= df['Profit'].mean()+ 3*df['Profit'].std()
Higher_limit
```

```python
Lower_limit= df['Profit'].mean()- 3*df['Profit'].std()
Lower_limit
```

```python
df['Profit']= np.where(
    df['Z-Score']>3, Higher_limit,
    np.where( df['Z-Score']<-3, Lower_limit, df['Profit'] )
)
```

```python
df.drop(columns='Z-Score', inplace=True)
```

##### 6.1.2 Applying IQR

On Amount, Quantity column IQR is applied because it is skewed distribution

###### Amount column

```python
aq1= df['Amount'].quantile(0.25)
aq3= df['Amount'].quantile(0.75)
```

```python
AIQR= aq3 - aq1
```

```python
aminrange= aq1- (1.5*AIQR)
amaxrange= aq3+ (1.5*AIQR)
```

```python
df['Amount']= np.where(
    df['Amount']<aminrange, aminrange,
    np.where(df['Amount']>amaxrange, amaxrange, df['Amount'])
)
```

###### Quantity column

```python
qq1= df['Quantity'].quantile(0.25)
qq3= df['Quantity'].quantile(0.75)
```

```python
QIQR= qq3 - qq1
```

```python
qminrange= qq1- (1.5*QIQR)
qmaxrange= qq3+ (1.5*QIQR)
```

```python
df['Quantity']= np.where(
    df['Quantity']<qminrange, qminrange,
    np.where(df['Quantity']>qmaxrange, qmaxrange, df['Quantity'])
)
```

#### 6.2 Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
features = ['Total_Amount', 'Total_Profit', 'Total_Quantity', 'Total_Orders']
```

```python
scaled_data = ss.fit_transform(customer_df[features])
```

# 

### 7. Optimal K Selection

#### Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
```

```python
K = range(1, 11)
```

```python
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(K, inertia, 'bo-')
plt.xticks(range(1, 11))
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(axis='x')
plt.show()
```

# 

### 8. K-Means Clustering

```python
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
customer_df['Cluster'] = kmeans.fit_predict(scaled_data)
```
